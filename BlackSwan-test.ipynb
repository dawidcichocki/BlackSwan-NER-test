```python
import spacy
```

```python
def NER(sentence, language_code):
    if language_code not in ['ca', 'zh', 'da', 'nl', 'en', 'fr', 'de', 'el', 'it', 'ja', 'lt', 'mk', 'xx', 'nb', 'pl', 'pt', 'ro', 'ru', 'es']:
        language_code = 'xx'
    model_string = f"{language_code}_core_web_sm" if language_code !='xx' else 'xx_ent_wiki_sm'
    ner_model = spacy.load(model_string, disable=["tok2vec", "tagger", "parser", "attribute_ruler", "lemmatizer"])
    output = ner_model(sentence)
    
    out = []
    for entity in output.ents:
        out.append({'text': entity.text,
                    'type': entity.label_, 
                    'start_pos': entity.start_char,
                    'end_pos': entity.end_char
                   })
    return out
```

```python
NER("American jazz orchestra", 'zz')
```
